# 矢量加法满足交换律

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义矢量 A 和 B
A = np.array([1, 2])
B = np.array([4, 2])

# 计算 A + B 和 B + A
A_plus_B = A + B
B_plus_A = B + A

# 创建图形和轴
fig, ax = plt.subplots()

# 绘制矢量 A, B, A + B 和 B + A
ax.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\overrightarrow{A}$') # 从原点开始画一个向量 A
ax.quiver(A[0], A[1], B[0], B[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\overrightarrow{B}$') # 从 A 点开始画一个向量 B
ax.quiver(0, 0, A_plus_B[0], A_plus_B[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.5, label=r'$\overrightarrow{A} + \overrightarrow{B}$') # 从原点开始画一个向量 A + B

ax.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='g', alpha=0.5) # 从原点开始画一个向量 B
ax.quiver(B[0], B[1], A[0], A[1], angles='xy', scale_units='xy', scale=1, color='r', alpha=0.5) # 从 B 点开始画一个向量 A
ax.quiver(0, 0, B_plus_A[0], B_plus_A[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.3, label=r'$\overrightarrow{B} + \overrightarrow{A}$') # 从原点开始画一个向量 B + A

# 在图中标出向量 A 和 B
ax.text(A[0]-0.2, A[1], r'$\overrightarrow{A}$', color='r', fontsize=12)
ax.text(B[0]+0.1, B[1]-0.2, r'$\overrightarrow{B}$', color='g', fontsize=12)

# 设置轴的范围
ax.set_xlim(0, 7)
ax.set_ylim(0, 7)

# 设置等比例
ax.set_aspect('equal')

# 添加网格和图例
ax.grid()
ax.legend()

# 显示图形
plt.show()
```

# 矢量加法满足结合律

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义矢量 A 和 B
A = np.array([1, 2])
B = np.array([4, 2])
C = np.array([-1, 1])

# 计算 A + B 和 B + C
A_plus_B = A + B
B_plus_C = B + C
A_plus_B_plus_C = A + B + C

# 创建图形和轴
fig, ax = plt.subplots()

# 绘制矢量 (A + B) + C 和 A + (B + C)
ax.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.5, label=r'$\overrightarrow{A}$') # 从原点开始画一个向量 A
ax.quiver(A[0], A[1], B[0], B[1], angles='xy', scale_units='xy', scale=1, color='g', alpha=0.5, label=r'$\overrightarrow{B}$') # 从 A 点开始画一个向量 B
ax.quiver(A_plus_B[0], A_plus_B[1], C[0], C[1], angles='xy', scale_units='xy', scale=1, color='y', alpha=0.5, label=r'$\overrightarrow{C}$') # 从 A + B 点开始画一个向量 C

ax.quiver(0, 0, A_plus_B[0], A_plus_B[1], angles='xy', scale_units='xy', scale=1, color='c', alpha=0.5, label=r'$\overrightarrow{A} + \overrightarrow{B}$') # 从原点开始画一个向量 A + B
ax.quiver(A[0], A[1], B_plus_C[0], B_plus_C[1], angles='xy', scale_units='xy', scale=1, color='m', alpha=0.5, label=r'$\overrightarrow{B} + \overrightarrow{C}$') # 从 A 点开始画一个向量 B + C
ax.quiver(0, 0, A_plus_B_plus_C[0], A_plus_B_plus_C[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\overrightarrow{A} + \overrightarrow{B} + \overrightarrow{C}$') # 从原点开始画一个向量 A + B + C


# 在图中标出向量 A 和 B
ax.text(A[0]-0.2, A[1]+0.1, r'$\overrightarrow{A}$', color='b', fontsize=12)
ax.text(A_plus_B[0]+0.1, A_plus_B[1]-0.2, r'$\overrightarrow{B}$', color='g', fontsize=12)
ax.text(A_plus_B_plus_C[0]-0.1, A_plus_B_plus_C[1]+0.1, r'$\overrightarrow{C}$', color='y', fontsize=12)

# 设置轴的范围
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)

# 设置等比例
ax.set_aspect('equal')

# 添加网格和图例
ax.grid()
ax.legend()

# 显示图形
plt.show()
```

# 点乘满足分配律

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 设置字体，防止中文乱码
rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
rcParams['axes.unicode_minus'] = False    # 解决负号无法正常显示的问题

# 定义矢量 A, B 和 C
A = np.array([5, 1])
B = np.array([1, 2])
C = np.array([3, 1])

# 计算 B + C
B_plus_C = B + C

# 计算点乘
dot_A_B_plus_C = np.dot(A, B_plus_C)
dot_A_B = np.dot(A, B)
dot_A_C = np.dot(A, C)

# 创建图形和轴
fig, ax = plt.subplots()

# 绘制矢量 A, B, C 和 B + C
ax.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\overrightarrow{A}$')
ax.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\overrightarrow{B}$')
ax.quiver(B[0], B[1], C[0], C[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\overrightarrow{C}$')
ax.quiver(0, 0, B_plus_C[0], B_plus_C[1], angles='xy', scale_units='xy', scale=1, color='m', label=r'$\overrightarrow{B} + \overrightarrow{C}$')

# 在图中标出矢量
ax.text(A[0], A[1]-0.5, r'$\overrightarrow{A}$', color='r', fontsize=12, fontweight='bold', style='italic')
ax.text(B[0]-0.1, B[1]+0.1, r'$\overrightarrow{B}$', color='g', fontsize=12, fontweight='bold', style='italic')
ax.text(B_plus_C[0]-0.4, B_plus_C[1]+0.1, r'$\overrightarrow{C}$', color='b', fontsize=12, fontweight='bold', style='italic')
ax.text(B_plus_C[0], B_plus_C[1], r'$\overrightarrow{B} + \overrightarrow{C}$', color='m', fontsize=12, fontweight='bold', style='italic')

# 计算 B 在 A 上的投影
proj_B_on_A = (np.dot(B, A) / np.dot(A, A)) * A

# 绘制 B 到投影的垂线段
ax.quiver(B[0], B[1], proj_B_on_A[0] - B[0], proj_B_on_A[1] - B[1], angles='xy', scale_units='xy', scale=1, color='g', alpha=0.5, label=r'$B$ 到投影的垂线段')

# 计算 C 在 A 上的投影
proj_B_plus_C_on_A = (np.dot(B_plus_C, A) / np.dot(A, A)) * A

# 绘制 C 到投影的垂线段
ax.quiver(B_plus_C[0], B_plus_C[1], proj_B_plus_C_on_A[0] - B_plus_C[0], proj_B_plus_C_on_A[1] - B_plus_C[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.5, label=r'$C$ 到投影的垂线段')

# 设置轴的范围
ax.set_xlim(0, 7)
ax.set_ylim(0, 7)

# 设置等比例
ax.set_aspect('equal')

# 添加网格和图例
ax.grid()
ax.legend()

# 显示图形
plt.show()
```

# 叉乘满足分配律

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 设置字体，防止中文乱码
rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
rcParams['axes.unicode_minus'] = False    # 解决负号无法正常显示的问题

# 定义矢量 B 和 C
B = np.array([1, 2])
C = np.array([3, 1])

# 计算 B + C
B_plus_C = B + C

# 定义旋转矩阵
theta = np.radians(90)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

# 旋转 B, C 和 B + C
B_rotated = np.dot(rotation_matrix, B)
C_rotated = np.dot(rotation_matrix, C)
B_plus_C_rotated = np.dot(rotation_matrix, B_plus_C)

# 创建图形和轴
fig, ax = plt.subplots()

# 绘制旋转前的矢量 B, C 和 B + C
ax.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='g', alpha=0.5, label=r'$\overrightarrow{B}$')
ax.quiver(B[0], B[1], C[0], C[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.5, label=r'$\overrightarrow{C}$')
ax.quiver(0, 0, B_plus_C[0], B_plus_C[1], angles='xy', scale_units='xy', scale=1, color='m', alpha=0.5, label=r'$\overrightarrow{B} + \overrightarrow{C}$')

# 绘制旋转后的矢量 B, C 和 B + C
ax.quiver(0, 0, B_rotated[0], B_rotated[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\overrightarrow{B}$')
ax.quiver(B_rotated[0], B_rotated[1], C_rotated[0], C_rotated[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\overrightarrow{C}$')
ax.quiver(0, 0, B_plus_C_rotated[0], B_plus_C_rotated[1], angles='xy', scale_units='xy', scale=1, color='m', label=r'$\overrightarrow{B} + \overrightarrow{C}$')

# 在图中标出旋转前的矢量
ax.text(B[0]-0.5, B[1]-0.2, r'$\overrightarrow{B}$', color='g', fontsize=12)
ax.text(B[0] + C[0]-0.5, B[1] + C[1]+0.1, r'$\overrightarrow{C}$', color='b', fontsize=12)
ax.text(B_plus_C[0]+0.2, B_plus_C[1]-0.2, r'$\overrightarrow{B} + \overrightarrow{C}$', color='m', fontsize=12)

# 在图中标出旋转后的矢量
ax.text(B_rotated[0]-1.7, B_rotated[1]-0.5, r'$\overrightarrow{A}\times\overrightarrow{B}$', color='g', fontsize=12)
ax.text(B_rotated[0] + C_rotated[0]-1.7, B_rotated[1] + C_rotated[1]-1, r'$\overrightarrow{A}\times\overrightarrow{C}$', color='b', fontsize=12)
ax.text(B_rotated[0] + C_rotated[0], B_rotated[1] + C_rotated[1]+0.1, r'$\overrightarrow{A}\times(\overrightarrow{B} + \overrightarrow{C})$', color='m', fontsize=12)

# 设置轴的范围
ax.set_xlim(-6, 10)
ax.set_ylim(0, 6)

# 设置等比例
ax.set_aspect('equal')

# 添加网格和图例
ax.grid()
ax.legend()

# 显示图形
plt.show()
```
