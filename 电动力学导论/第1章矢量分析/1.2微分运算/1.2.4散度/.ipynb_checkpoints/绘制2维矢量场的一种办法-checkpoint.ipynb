# 绘制二维矢量场的一种方法

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rcParams
# 设置字体，防止中文乱码
rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
rcParams['axes.unicode_minus'] = False    # 解决负号无法正常显示的问题

# 定义矢量场函数
def vector_field(x, y):
    # 这里定义矢量场 A = (A_x, A_y)，可以修改为你需要的函数
    A_x = -y  # x 方向分量
    A_y = x   # y 方向分量
    return A_x, A_y

# 创建网格点
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# 计算每个网格点的矢量场分量
A_x, A_y = vector_field(X, Y)

# 绘制矢量场
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, A_x, A_y, angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.xlabel('X 轴')
plt.ylabel('Y 轴')
plt.title('二维矢量场图像')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()
```
